# Conclusiones

## Evidencia (qué mostraron los datos)
- El consumo crece con el plan (mediana 558 → 833 → 1081).
- Edad, inactividad y tickets no se asocian con el consumo (r≈0).
- El PCA no reduce dimensionalidad (4 componentes ~25% c/u) ni separa perfiles.

## Interpretación (qué significa)
- El único factor que estructura el comportamiento es el plan contratado.
- Las demás variables son estadísticamente independientes: el comportamiento del usuario es esencialmente aleatorio respecto de ellas.

## Conclusión / recomendación (qué decisión habilita)

- **No segmentar campañas por edad:** la edad no se asocia con el consumo, así que personalizar contenido por esa variable no tendría efecto.
- **El consumo es el diferenciador real entre planes:** los usuarios Premium consumen casi el doble que los Básico. Conviene apalancar ese dato en la propuesta de valor (el plan caro efectivamente se usa más).

## Limitaciones
- Dataset con alta suciedad original e imputaciones que aplanan diferencias finas.
- 472 usuarios sin fecha de login excluidos del PCA.
- La ausencia casi total de correlaciones sugiere que el dataset podría ser sintético, y limita la generalización de las conclusiones a usuarios reales.
